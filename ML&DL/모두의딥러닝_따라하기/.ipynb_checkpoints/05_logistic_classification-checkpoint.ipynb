## logistic Classfication 구현하기

```python
import tensorflow as tf
```

```python
# 데이터 구현

x_data = [[1,2],[2,3],[3,1],[4,3],[5,3],[6,2]]
y_data = [[0],[0],[0],[1],[1],[1]] # binary : 0 또는 1

# feed_dict를 사용하기 위함. None인 이유는 데이터가 몇개 들어와도 상관이 없기 때문이다.
X = tf.placeholder(tf.float32, shape = [None, 2])
Y = tf.placeholder(tf.float32, shape = [None, 1])
```

```python
# y가 한개로 나가므로 1개
w = tf.Variable(tf.random_normal([2,1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')
```

```python
h = tf.sigmoid(tf.matmul(X, w) + b)
```

```python
cost = -tf.reduce_mean(Y*tf.log(h)+(1-Y)*tf.log(1-h))
```

```python
train = tf.train.GradientDescentOptimizer(learning_rate = 0.01).minimize(cost)
```

```python
# 예측 값이 성공인지 아닌지 보자
# 0.5 보다 작으면 fail 크면 pass로 하겠다.
# cast는 true면 1, false면 0을 반환한다. 타입에 맞게 반환해준다.
predicted = tf.cast(h > 0.5, dtype = tf.float32)
# 10개중 7개 맞으면 70퍼..
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype = tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        cost_val, _ = sess.run([cost, train], feed_dict = {X : x_data, Y: y_data})
        if step % 500 == 0:
            print(step, cost_val)
            
    h, c, a = sess.run([h, predicted, accuracy], feed_dict = {X : x_data, Y: y_data})
    
    print(h, c, a)
```

실제 데이터에 적용해보자

```python
import numpy as np
```

```python
xy = np.loadtxt('./05diabetes.csv', delimiter = ',', dtype = np.float32)
xdata = xy[:,0:-1]
ydata = xy[:,[-1]]
# 1이면 당뇨다.
```

```python
x = tf.placeholder(tf.float32, shape = [None, 8])
y = tf.placeholder(tf.float32, shape = [None, 1])

w = tf.Variable(tf.random_normal([8,1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')
```

```python
h = tf.sigmoid(tf.matmul(x, w)+b)
```

```python
cost = -tf.reduce_mean(y*tf.log(h)+(1-y)*tf.log(1-h))
train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)
```

```python
predicted = tf.cast(h > 0.5, dtype = tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, y), dtype = tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        sess.run(train, feed_dict = {x : xdata, y: ydata})
        if step % 500 == 0:
            print(step, sess.run(cost,feed_dict = {x : xdata, y: ydata} ))
            
    h, c, a = sess.run([h, predicted, accuracy], feed_dict =  {x : xdata, y: ydata})
    
    print(h, c, a)
```
