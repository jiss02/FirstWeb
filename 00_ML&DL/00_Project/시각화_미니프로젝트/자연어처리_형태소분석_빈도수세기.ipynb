```python
# 파일 불러와서 소문자로 바꿔준다
document_text = open('presentation.txt', 'r')
text = document_text.read().lower()

# 검열 안되는 문자 따로 지워주기
text = text.replace('_',' ')
```

```python

# 정규표현식에 맞게 단어를 토큰으로 만들어 준다. 
import nltk 
nltk.download('punkt')
from nltk.tokenize import sent_tokenize
from nltk.tokenize import RegexpTokenizer
retokenize = RegexpTokenizer("[\w]+")
data = retokenize.tokenize(text)
```

```python
# 불용어 처리

import nltk
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 

stop_words = set(stopwords.words('english')) 

result = [] 

for w in data: 
    if w not in stop_words: 
        result.append(w) 

#print(result) 결과 확인해보기
```

```python
# 형용사인지 동사인지 태그를 붙여줌. 튜플형식
nltk.download('tagsets')
nltk.download('averaged_perceptron_tagger')
from nltk.tag import pos_tag
tagged_list = pos_tag(data)
```

```python
# 의미있는 품사들을 넣어주세영
nouns_list = [t[0] for t in tagged_list if t[1] == "NN"]

# 여기다가!
verb=["VB","VBP", "VBG","VBD","VBZ", "VBN", "VBT", "RB", "JJ"]
Verb_list = [t[0] for t in tagged_list if t[1] in verb]
```

```python
#print(Verb_list)
#print(nouns_list)
```

```python
# 딕셔너리에 넣어주자
words_count={}

re = Verb_list + nouns_list

# 쓰레기통 ㅇㅅㅇ
gar = ["was", "were", "did", "didn", "don", "have", "had", "are"]

# 글자수가 3개이상인것만, 쓰레기통에 없는 것만 딕셔너리에 넣는다!
for word in re:
    if (len(word) >= 3 and word not in gar):
        if word in words_count:
            words_count[word] += 1
        else:
            words_count[word] = 1

# 여긴 보지마세영            
# sorted_words = sorted([(k,v) for k,v in words_count.items()], key=lambda word_count: -word_count[1])
# print([w[0] for w in sorted_words])
# print(sorted(words_count, key=lambda k : words_count[k], reverse=True))
# print(words_count)
```

```python
# 프레임에 넣기
import pandas as pd
df = pd.DataFrame(list(words_count.items()), columns=['words', 'Freq'])
```

```python
# 빈도수가 5이상인것만 뽑는다. 내림차순 정렬은 덤!
final = df.sort_values('Freq',ascending=False)
final = final[final['Freq']>=5]

final
```
