## 텐서플로로 간단한 선형회귀 구현

costfunction을 최소화하는 것이 목표이다.

```python
import tensorflow as tf
```

```python
# 가설 

x_train = [1,2,3]
y_train = [1,2,3]

# 텐서플로가 사용하는 variable. 학습하는 과정에서 스스로 변형시킨다.
w = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

hypothesis = x_train * w + b
```

```python
# 코스트
cost = tf.reduce_mean(tf.square(hypothesis - y_train))
```

```python
# 코스트 최소화를 경사하강법으로 구해보자
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
```

```python
# session을 통한 실행
sess = tf.Session()
# variables를 사용하기 전에는 무조건 초기화해주어야한다.
sess.run(tf.global_variables_initializer())
```

```python
# trian을 실행시킨 다는 것은 코스트를 실행시키는 것이며 이는 가설을 실행시키는 것이고, 가설의 w와b까지 실행을 시키는 것과 같다.
for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost), sess.run(w), sess.run(b))
```

## placeholder, variables

__플레이스홀더__

placeholder는 처음에 정의하지않고, 나중에 사용자가 넣어줄 수 있는 것이다.

모델을 만들어 놓고 나중에 값을 바꿔가면서 줄 수 있는 것이다.

shape을 파라미터로 줄 수 있는데,`[none]`을 주게되면 제한없이 줄 수 있다.

```python
# x와 y를 정의하지 않고 사용해보겠다
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

for step in range(2001):
    # 한꺼번에 실행시켜서 변수에 대응이 가능하다.
    cost_val, w_val, b_val,_ = \
        sess.run([cost,w,b, train], feed_dict={X: [1,2,3], Y:[1,2,3]})
    
    if step % 20 == 0:
        print(step, cost_val, w_val, b_val)
```
